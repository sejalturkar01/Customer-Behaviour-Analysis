```python
# think about how should we replace the review rating missing values as this columns has review accoording to category like clothing,jewellery 
#so we will replace missing values with the indivudual categorial median.
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# this is snake caseing makes it more easier to read column name and also do not give errors 
#customer_id
```

```python
df.columns = df.columns.str.lower() # converting all the columns to lower case
df.columns = df.columns.str.replace(' ','_') #replacing the space with underscore
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) # converting the columns to easier name
```

```python
df.columns
```

```python
#creating a column age_group
labels = ['young_adults' , 'adults' , 'Middle_age' , 'Senior']
df['age_groups'] = pd.qcut(df['age'], q=4, labels = labels)
#qcut :it cuts the age into 4 parts and labels them according to the labels given
```

```python
df[['age','age_groups']].head(10)
```

```python
# creating column purchase frequency rate
# converting the frequency from text to number for better analysis
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

# map function is used to convert text into number

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine

# SQLite - no server needed, creates a file automatically
engine = create_engine('sqlite:///customer_behavior.db')

# Write DataFrame to SQLite
df.to_sql('customer', engine, if_exists='replace', index=False)

# Read back
import pandas as pd
result = pd.read_sql("SELECT * FROM customer LIMIT 5;", engine)
print(result)
```

```python
pd.read_sql("""Select COUNT(customer_id)as total_customer,Round(AVG(purchase_amount),2)as avg_purchase_product,SUM(purchase_amount) as total_revenue,subscription_status from customer group by subscription_status order by total_revenue,avg_purchase_product desc""",engine)
```

```python
pd.read_sql(""" select item_purchased,Round(100 * SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END)/COUNT(*),2) as discount_rate from customer group by item_purchased order by discount_rate desc limit 5""",engine)
```

```python
pd.read_sql("""
    WITH customer_type AS (
        SELECT 
            customer_id,
            previous_purchases,
            CASE 
                WHEN previous_purchases = 1 THEN 'New'
                WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
                ELSE 'Loyal'
            END AS customer_segment
        FROM customer
    )
    SELECT 
        customer_segment,
        COUNT(*) AS number_of_customers
    FROM customer_type
    GROUP BY customer_segment
    ORDER BY number_of_customers DESC
""", engine)
```

```python
pd.read_sql("""WITH item_counts AS (
    SELECT category,
           item_purchased,
           COUNT(customer_id) AS total_orders,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(customer_id) DESC) AS item_rank
    FROM customer
    GROUP BY category, item_purchased
)
SELECT item_rank,category, item_purchased, total_orders
FROM item_counts
WHERE item_rank <=3; """,engine)
```

```python
pd.read_sql(""" SELECT subscription_status,
       COUNT(customer_id) AS repeat_buyers
FROM customer
WHERE previous_purchases > 5
GROUP BY subscription_status;""",engine)
```

```python
pd.read_sql("""
    SELECT age_groups,
           SUM(purchase_amount) AS total_revenue,
           COUNT(*) AS total_customers,
           ROUND(AVG(purchase_amount), 2) AS avg_spend
    FROM customer
    GROUP BY age_groups
    ORDER BY total_revenue DESC
""", engine)
```
